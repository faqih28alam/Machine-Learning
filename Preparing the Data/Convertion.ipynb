### i. Sisipkan matriks y' di kolom ketiga dari x, simpan di z2!

```python
print("matrix y'")
print(yt)
print("matrix x")
print(x)
print("hasil sisipan matrix y' di kolom ke 3 matrix x")
z2 = np.concatenate((yt,x),axis=1)
print(z2)
```

### j. Tambahkan angka 10 pada baris ketiga dari p, simpan di p1!

### k. Sisipkan p1' pada baris pertama dari z2, simpan di z3!
